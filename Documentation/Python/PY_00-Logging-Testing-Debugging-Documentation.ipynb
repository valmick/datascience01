# Logging


````
import logging
````

````
logging.basicConfig(level=logging.DEBUG)
````

````
def fib(n):
  if n == 0 or n == 1:
    return 1
  else:
    return fib(n-1)+fib(n-2)
````

````
fib(1)
````

````
fib(2)
````

````
fib(5)
````

````
def fib(n):
  logging.debug(n)
  if n == 0 or n == 1:
    return 1
  else:
    return fib(n-1)+fib(n-2)
  
````

````
fib(6)
````

````
# Disable logging
logging.disable(logging.CRITICAL) # it seems once this is executed, it sets the variable permanently
````

````
fib(6)
````

# Testing

Reference: https://docs.python.org/2/library/doctest.html

````
import doctest

````

````
def fib3(n):
  """
  >>> fib(5)
  8
  >>> fib(1)
  1
  """
  if n == 0 or n == 1:
    return 1
  else:
    return fib3(n-1)+fib3(n-2)
````

````
doctest.run_docstring_examples(fib)
````
